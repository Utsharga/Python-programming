```python
#sigma notation
total = 0
for i in range(11):
    #total += i*2
    total = total + i*2
    print (total)
print(total)
```

```python
#product notation
total = 1
for i in range(1,11):
    #total = total*3*i
    total *= 3*i
    print (total)
print (total)
```
